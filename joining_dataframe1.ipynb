```python
data4 = {'emp_id':[101,102,103,104],
        'salary':[234,345,200,400],
        'pf':[23,34,56,78]}
```

```python
data5 = {'emp_id':[101,102,103,104],
        'mobile_no':[3434,5656,767,678],
        'house_no':[345,345,345,345]}
```

```python
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
data6 = {'emp_id_1':[101,102,103,104],
        'mobile_no_1':[3433,5656,767,678],
        'house_no_1':[345,345,345,345]}
```

```python
data7 = {'emp_id_2':[101,102,103,104],
        'mobile_no_2':[3433,5656,767,678],
        'house_no_2':[345,345,345,345]}
```

```python
df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)
```

```python
# different column names
pd.merge(df6,df7,left_on='emp_id_1',right_on='emp_id_2')
```

```python
data8 = {'emp_id':[101,102,103,104],
        'salary':[234,345,200,400],
        'house_no':[23,34,56,78]}

data9 = {'emp_id':[101,102,104,105,123],
        'salary':[234,345,200,878,43],
        'house_no':[345,345,345,345,345]}
```

```python
df8 = pd.DataFrame(data8)
df9 = pd.DataFrame(data9)
```

#  Join

```python
pd.merge(df8,df9,on=['emp_id','salary'])
```
