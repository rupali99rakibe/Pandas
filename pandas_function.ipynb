```python
import pandas as pd
```

```python
df = pd.read_csv("D:\practice_data_set\car_data.csv")
df.head()
```

```python
for i in df['Present_Price']:
    if i > 5:
        print("High")
    else:
        print('Low')
```

```python
def review(x):
    if x > 5:
        return 'High'
    else:
        return 'Low'
```

```python
review(5.59)
```

```python
df['review'] = df['Present_Price'].apply(review)
```

```python
df.head()
```

```python
df['Length_of_car'] = df['Car_Name'].apply(lambda x : len(x))
```

```python
df['Km'] = df['Kms_Driven'].apply(lambda x : x/5)
```
